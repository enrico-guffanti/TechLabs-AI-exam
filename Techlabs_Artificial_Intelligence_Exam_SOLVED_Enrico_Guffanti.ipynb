### Your coding skills have gained you a job as an options trader at a successful hedge fund! Congratulations!!

At the first day, your boss comes to you and asks you, whether he should buy a _call option_* with a certain set of characteristics for 1€.

    *A call option gives you the right (but not obligation) to buy a share for a certain strike price. In other words, if the stock price is higher than the strike price, you get the difference, otherwise, you get 0: 
    
    callpayoff = max(stockprice - strikeprice, 0) 

To price the option, you shall build a monte-carlo simulator which generatates _1 000 000_ random walks, each representing the stock price in one year, which is when the option can be expired. By taking the average of these payouts, you will get the expected payout at expiry!

Luckily, your boss has also given you the characteristics and hints for how the stock price moves: 

The stock price follows a student T distribution with 3 degrees of freedom (the rvs function within the t class from scipy.stats package is a great tool for creating random walks with this distribution: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html  ctrl+f "rvs") 

- The stock price today: 69
- Stock returns follow student T distribution with 3 degrees of freedom
- drift of 10% per year 
- volatility of 20%
- strike price of the option is 96

### Build a function that returns the price of the call option, and shows your boss what payout he can expect from the option in order to explain him, whether he should buy it or not.

    P.S.: If you run into trouble on your way, you can always ask for tips from your boss and since he is a nice guy, he will also give a good grade for all attempts!

```python
#Definition of variables
price_call_option = 1
p_0 = 69
p_strike = 96
degrees_of_freedom = 3
distribution_type = "student T"
days = 365
drift = 0.1
drift_daily = 0.1/365
volatility = 0.2
number_of_simulations = 1000000
```

```python
#Importing relevant libraries
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import t

#Defining function to return future prices "periods_from_now" in the future
def price_generator(p_today, dr_d, vol, periods_from_now):
  rand_error = t.rvs(df=degrees_of_freedom, size=periods_from_now)
  p_list = [p_0]
  for i in range(periods_from_now-1):
    p_today = p_today + dr_d*p_today + rand_error[i]*vol*p_today
    if p_today >= 0:
      p_list = p_list + [p_today]
    else:
      p_list = p_list + [0]
  return p_list
```

```python
#Simulation of number_of_simulations random walks
def random_walks_simulator(reps):
  walks = []
  for j in range(reps):
    walk = price_generator(p_today=p_0, dr_d=drift_daily, vol=volatility, periods_from_now=days)
    walks = walks + [walk]
  return walks
```

```python
#Extracting stock price on the last day of the year and taking the average
scenarios = random_walks_simulator(reps=number_of_simulations)
closing_prices = [item[-1] for item in scenarios]
avg_p_1 = np.mean(closing_prices)
avg_p_1
```

```python
#Defining profit scenarios
profit_NO_opt = avg_p_1 - p_0 #Profit in case the option is NOT purchased
profit_opt = avg_p_1 - min(avg_p_1, p_strike) - price_call_option #Profit in case the option is purchased
profits = [profit_NO_opt, profit_opt]
```

```python
#Deciding on what to do based on profits
Decision = max(profits)
index_max_in_profits = profits.index(Decision)
a = [Decision, index_max_in_profits]

b = ["don't buy the option", "buy the option"]

c = ""
if a[0]==abs(price_call_option):
  c = "the forecasted future price is lower than strike price: do not use the right granted by the option."

print("In order to maximize profit, you should select the option with index "+str(a[1])+" from the profits list. Therefore, what you have to do is "+str(b[a[1]])+".\nHowever, the forecasted future price is lower than strike price: do not use the right granted by the option. Thus, the payoff you will get following this first best route is "+str(a[0]))
```
